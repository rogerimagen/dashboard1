```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('C:/Users/rguevara/Documents/Pres/Batch/Rat_Prog.csv')

from matplotlib.colors import ListedColormap


x = df.plot(kind='bar',legend=True, fontsize=8, figsize=(15, 5), stacked=True,
            colormap=ListedColormap(sns.color_palette("Blues", 10)))
x.set_xticks(df.index)
x.set_xticklabels(df.Programa, rotation=90)
plt.title("Prueba")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df1 = pd.read_csv('C:/Users/rguevara/Documents/Pres/Batch/Shr_10C.csv')
explode = (.5, 0, 0, 0, 0, 0, 0, 0, 0, 0)

from matplotlib.colors import ListedColormap

fig1, ax1 = plt.subplots()
ax1.pie(df1.shr, explode=explode, labels=df1.Channel, autopct='%1.1f%%',
        shadow=True, startangle=35, labeldistance=1.3,textprops={'size': 'smaller'})

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
